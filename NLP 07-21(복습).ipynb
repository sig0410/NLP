```python
def read_data(filename, encoding = 'utf-8'):
    with open(filename,'r', encoding=encoding) as f:
            data = [line.split(' ') for line in f.read().splitlines()]
    return data

data = read_data(filepath + 'wiki202007_nationalcorpus_naverratings_morphed.txt', 'utf-8')

# print(len(data))
# print(data[:3]) #구동확인

from gensim.models import Word2Vec
os.chdir(filepath + 'word2vec')

model = Word2Vec(sentences= data, size = embedding_dim, window= 10, min_count=5, workers= 4, sg=1)
#위에서 형태소분석한 단어들을 바탕으로 임베딩 레이어는 50개 최소단어빈도는 5, input은 data의 크기로 지정한 skipgram
model.save('embedding_window10_mincnt5_skipgram.model1')
print('모델 완')
```

```python
import os 
naver_dir = '/Users/Moon/Desktop/Summer/NLP/pytest/네이버뉴스-생활문화_다중/'
train_dir = os.path.join(naver_dir, 'train')
test_dir = os.path.join(naver_dir, 'test')

label_types = ['건강정보', '공연전시', '도로교통', '여행레저', '음식맛집', '자동차시승기', '종교', '책', '패션뷰티']
#os패키지를 이용하여 디렉토리를 지정 
```

```python
def make_texts_labels(dir, label_types, encode = 'utf-8'):
    texts =[]
    labels = []
    #각각 빈리스트를 만든다
    
    for label_type in label_types:
        dir_name = os.path.join(dir, label_type) 
        #각각의 라벨폴더에 접근
        for fname in os.listdir(dir_name):
            if fname[-4:] == '.txt': #끝에서부터 4번째까지가 .txt일때 
                f = open(os.path.join(dir_name,fname), encoding = encode)
                texts.append(f.read())
                f.close()
                
                if label_type == label_types[0]: #label_types[0]이 저장되어있는 것과 같을때 
                    labels.append(0)
                elif label_type == label_types[1]:
                    labels.append(1)
                elif label_type == label_types[2]:
                    labels.append(2)
                elif label_type == label_types[3]:
                    labels.append(3)
                elif label_type == label_types[4]:
                    labels.append(4)
                elif label_type == label_types[5]:
                    labels.append(5)
                elif label_type == label_types[6]:
                    labels.append(6)
                elif label_type == label_types[7]:
                    labels.append(7)
                elif label_type == label_types[8]:
                    labels.append(8)
    return texts,labels
#os패키지에 대해 이해가 부족함
#특정 폴더에 있는 데이터를 불러와서 빈리스트에 저장하고 해당 폴더의 이름을 확인하여 labels리스트에 분류를 적는 함수를 만듬 
```

```python
texts, labels = make_texts_labels(train_dir, label_types)

print(texts[0])
print(len(texts))
print(labels[0])
print(len(labels))
```

```python
texts = [rhinoMorph.onlyMorph_list(rn, sentence, pos= ['NNG', 'NNP','NP', 'VV', 'VA', 'XR', 'IC', 'MM', 'MAG', 'MAJ'], eomi = False) for sentence in texts]
#texts를 sentence라는 반복문으로돌리고 해당 반복문에서 생성된 객체를 형태소분석을 진행해서 다시 texts로 저장 
print(texts[0])
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
import math
```

```python
validation_ratio = math.floor(len(texts) * 0.3)
max_words = 10000
maxlen = 200
class_number = 9
```

```python
tokenizer = Tokenizer(num_words = max_words)
#max_words로 지정한만큼 토큰화 
tokenizer.fit_on_texts(texts)
#texts를 토큰화해줌 
word_index = tokenizer.word_index
#word_index는 토큰화한것에 대한 각 단어의 인덱스를 저장 
```

```python
print('전체 %s개의 고유한 토큰을 찾았습니다.' %len(word_index))
print(word_index)
```

```python
data = tokenizer.texts_to_sequences(texts)
#각 단어 인덱스를 숫자 리스트로 변환해줌 
print(len(data))
#전체 데이터셋이 90개임으로 잘된것을 확인 
```

```python
data = pad_sequences(data, maxlen = maxlen)
print(data)
print(data[0])
#한 리스트당 길이를 200개로 지정을 해줌 
```

```python
import numpy as np
```

```python
def to_one_hot(labels, dimension):
    results = np.zeros((len(labels), dimension))
    #0행렬을 만듬 
    for i,label in enumerate(labels):
        results[i,label] = 1.
    return results
#위의 매개함수의 뜻은 label이 0이면 i가 label의 길이만큼 증가하고 이후 label이 1이 되는 과정을 계속 반복 
```

```python
labels = to_one_hot(labels, dimension = class_number)
#잘못누름 labels에 저장됨
```

```python
labels
#정확하게 출력이 되는 것을 확인을 할 수 있음
#한 리스트의 값을 전부 더하면 1이 되는 softmax함수를 사용
#softmax함수는 output이 확률의 형태로 나타나며 전부 더했을때 1이됨
```

```python
indices = np.arange(data.shape[0])
np.random.shuffle(indices)
data = data[indices]
labels = labels[indices]
#아마 섞어주는 역활일거임 
#나중에 다시 보기!!😊
```

```python
X_train = data[validation_ratio:]
y_train = data[validation_ratio:]
X_test = data[:validation_ratio]
y_test = data[:validation_ratio]
#validation_ratio를 기준으로 train과 test셋을 자름 
```

```python
filepath
```

```python
embeddings_index = {}
f = open(os.path.join(filepath + 'word2vec', 'embedding_window10_mincnt5_skipgram.txt'),
        encoding = 'utf-8')
for line in f:
    values = line.split()#행을 분리
    word = values[0]#단어는 value의 첫번쨰에 존재 
    coefs = np.asarray(values[1:], dtype = 'float32')#첫번쨰이후부터는 임베딩값만 존재 
    embeddings_index[word]= coefs
f.close()
```

```python
embedding_matrix = np.zeros((max_words, embedding_dim))
#0으로 채워진 빈행렬을 만듬 
for word, i in word_index.items():
    if i < max_words:
        embedding_vector = embeddings_index.get(word) #해당 단어의 벡터를 가져옴 
        if embedding_vector is not None:
            embedding_matrix[i] = embedding_vector 
#이부분 이해가 안감 
#다시 보기
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers
```

```python
model = models.Sequential()#모델 생성 
model.add(layers.Embedding(input_dim = max_words, output_dim = embedding_dim, input_length = maxlen))
#인풋은 10000개 output은 50개 각 input의 길이는 200
model.add(layers.Flatten())
model.add(layers.Dense(units =32 , activation = 'relu'))
model.add(layers.Dense(units = class_number, activation = 'softmax'))
model.summary()
```

```python
model.layers[0].set_weights([embedding_matrix])
model.layers[0].trinable = False
#layers[0]에는 단어라서 학습이 안되게 하는건가???....
#이까지가 keras 임베딩의 과정 
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer = 'rmsprop', metrics = ['acc'])
```

```python
history = model.fit(X_train, y_train, epochs = 10, batch_size = 64)
history_dict = history.history
```
