```python
def read_data(filename, encoding = 'cp949'):
    with open(filename, 'r', encoding = encoding ) as f:
        data =[line.split("\t") for line in f.read().splitlines()]
        data = data[1:]
    return data
  #읽기 함수
```

```python
def write_data(data, filename, encoding = 'cp949'):
    with open(filename, 'w', encoding = encoding ) as f:
        f.write(data)
# 쓰기함수
```

### 감성분석

1. 감성점수 > 0 긍정 
2. 감성점수 < 0 부정 
3. 감성점수 = 0 중립

```python
positive = read_data('/Users/Moon/Desktop/빅데이터 청년인재/NLP/pytest/positive.txt')
negative = read_data('/Users/Moon/Desktop/빅데이터 청년인재/NLP/pytest/negative.txt')
```

```python
def cntWordInLine(data, senti):
    senti_found = []
    for onedata in data:
        oneline_word = onedata.split(' ')#공백단위로 분리 
        senti_temp = 0 #기본값 0으로 지정 
        for sentiword in senti:
            if sentiword[0] in oneline_word:#공백단위로 분리한 단어가 있으면 senti_temp의 값을 1씩올림 
                senti_temp += 1 
        senti_found.append(senti_temp)#senti_found에 저장 
    return senti_found #최종적으로 senti_found를 리턴 
```

```python
data_senti_poscnt = cntWordInLine(data_text, positive)
#긍정인 단어를 파악
data_senti_negcnt = cntWordInLine(data_text, negative)
```

```python
print(data_senti_poscnt[:30])
#해당단어가 positive데이터안에 있었던 횟수를 보여줌 

print(data_senti_negcnt[:30])
#해당단어가 negative데이터안에 있었던 횟수를 보여줌 
```

```python
import pandas as pd
```

```python
newdata = pd.DataFrame({'id' : data_id, 'text' : data_text, 'original' : data_senti,
                        'pos' : data_senti_poscnt, 'neg' : data_senti_negcnt})
#새로운 데이터프레임을 만들어주고 새롭게 만든 컬럼들도 추가 

senti_score = newdata['pos'] - newdata['neg']
#긍정을 찍힌 개수에서 부정으로 찍힌 개수를 뺴줌 
newdata['senti_score'] = senti_score
#새로운 컬럼을 만들어줌
```

```python
newdata.loc[newdata.senti_score > 0, 'new' ] = 1
newdata.loc[newdata.senti_score <= 0, 'new'] = 0
#0을초과하면 1, 0이하면 0으로 맵핑 
```

```python
newdata.loc[pd.to_numeric(newdata.original) == newdata.new, 'matched'] = 'True'
newdata.loc[pd.to_numeric(newdata.original) != newdata.new, 'matched'] = 'False'
```

```python
score = newdata.matched.str.count('True').sum() / (newdata.matched.str.count('True').sum() + 
                                                   newdata.matched.str.count('False').sum())
#T와 F를 더한 값을 분모에 T만인 값을 분자에 두고 score를 도출  
print(round(score,3)*100,'%')
```

### 텍스트 이진분류

```python
validation_ratio = math.floor(len(data)*0.3)
max_words = 10000
maxlen = 200
tokenizer = Tokenizer(num_words = max_words) #빈도가 많은 순으로 max_words만큼 뽑아냄
tokenizer.fit_on_texts(data_text)
word_index = tokenizer.word_index #단어 인덱스를 구성
print(len(word_index))
print(word_index)#토큰화한 단어에 고유한 인덱스번호를 부여 
```

```python
d = pad_sequences(d,maxlen = maxlen)
print(len(d[0]))
#maxlen에서 지정한것이 200이며 제대로 padding이 된것을 알 수 있다
print(d[0])
```

```python
def to_one_hot(sequences, dimension):
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        results[i, sequence] = 1. 
    return results
```

```python
d1 = to_one_hot(d, dimension = max_words)
```

```python
labels = np.asarray(data_senti).astype('float32')
```

```python
print(len(d1[0]))
#원 핫 인코딩이 잘변환된것을 확인 
```

```python
indices = np.arange(d1.shape[0])
np.random.shuffle(indices)
data = d1[indices]
labels = labels[indices]
#197513까지의 숫자를 생성하여 랜덤으로 섞음 
```

```python
X_train = data[validation_ratio:]
y_train = labels[validation_ratio:]

X_val = data[:validation_ratio]
y_val = labels[:validation_ratio]
#train set안에서 train데이터와 validation데이터를 분리 
#추후 학습과정에서 잘학습되고 있는지를 확인하기 위해 validation을 사용 
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
#모델을 새로 정의 
#Sequential은 층을 순서대로 쌓아올리는 방식
```

```python
model.add(Dense(64, activation = 'relu', input_shape = (max_words,)))
#첫번쨰 은닉층이며 64개의 은닉노드로 구성되어 있고 마지막에 output으로 나가기전에 relu함수를 통해 값을 도출함
#input_shape : input을 지정 
model.add(Dense(32, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))
#마지막은 1개를 출력함 이진분류라서 1로 지정하였으며 sigmoid함수를 통해 최종결과를 확률형태로써 보여줌 
model.summary()
#Dense함수를 통해 각 층의 옵션을 구체적으로 결정할 수 있음
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'rmsprop', metrics = ['acc'])
history = model.fit(X_train, y_train, epochs = 10, batch_size = 32, validation_data = (X_val, y_val))
#train셋으로 모델을 학습하고 validaiton으로 학습과정을 확인할 수 있음
#총 10회의 학습을 진행하며 한번에 32개씩 학습을 하고 잇음 
```
